```python
import numpy as np
import matplotlib.pyplot as plt


values =[ 10, 500, 1000, 5000, 10000, 50000, 80000, 100000, 200000]
b = []
for i in values:
    x = np.random.uniform(low=0, high=1, size=[i, 1])
    y = np.random.uniform(low=0, high=1, size=[i, 1])
    inside = (x-0.5)**2+(y-0.5)**2<0.25
    pi = 4*np.sum(inside)/i
    b.append(4*np.sum(inside)/i)
    print('if i value is {}, then estimated Pi value is: {}'.format(i, pi))
    x_in = x[inside]
    y_in = y[inside]
    fig=plt.figure(figsize=(4,4))
    square_plot = fig.add_subplot(1, 1, 1)
    s=plt.Rectangle((0,0), 1, 1,fill=False)
    c=plt.Circle((0.5,0.5),radius=0.5,fill=False)
    square_plot.add_patch(s)
    square_plot.add_patch(c)
    plt.scatter(x, y, s=1)
    plt.scatter(x_in, y_in, color='red', s=1)
    plt.show()



y_axis = np.array(b)
x_axis=np.array(values)
plt.xlabel('Values')
plt.ylabel('estimated Value of Pi')
plt.title('difference of Accuracy of Pi')
plt.plot(x_axis,y_axis)
plt.show()
```
