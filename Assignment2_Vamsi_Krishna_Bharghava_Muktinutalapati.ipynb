```python



def continent_country(country):
    if country in Africa:
        return "Africa"
    elif country in Asia:
        return "Asia"
    elif country in Europe:
        return "Europe"
    elif country in NorthAmerica:
        return "NorthAmerica"
    elif country in SouthAmerica:
        return "SouthAmerica"
    elif country in Australia :
        return "Australia"
    else:
        return "Oceania"
df1["continent"]=df1["country"].apply(lambda x:continent_country(x))
df1
```

Write a Python program to find the country/region in each continent that has the highest average HIV estimated prevalence of people ages from 15 to 49 of from year 2000 to 2011.

```python
#Filtering data from 2000 to 2011
df2=df1[['country','continent',2000,2001,2002,2003,2004,2005,2006,2007,2008,'2009','2010','2011']]
df2
```

```python
#Finding highest average HIV estimated prevalence of people ages from 15 to 49 of from year 2000 to 2011.

df2['Highest HIV Estimated prevelance']= df2.iloc[:, 2:14].mean(axis=1)
df2
```

```python
df3=df2.sort_values(by='continent',ascending=True)
df3
```

```python

df4=pd.DataFrame(df3.groupby(["continent","country"])["Highest HIV Estimated prevelance"].max())
df4
```

```python
highhiv_average = df4.groupby('continent')['Highest HIV Estimated prevelance'].idxmax()
highhiv_average=df4.loc[highhiv_average]
highhiv_average=highhiv_average.reset_index()
highhiv_average
```

Find the country/region in each continent that has the lowest average HIV estimated prevalence of people ages from 15 to 49 of from year 2000 to 2011.

```python
df5=pd.DataFrame(df3.groupby(["continent","country"])["Highest HIV Estimated prevelance"].min())
df5
```

```python
lowhiv_average = df4.groupby('continent')['Highest HIV Estimated prevelance'].idxmin()
lowhiv_average=df4.loc[lowhiv_average]
lowhiv_average=lowhiv_average.reset_index()
lowhiv_average
```
